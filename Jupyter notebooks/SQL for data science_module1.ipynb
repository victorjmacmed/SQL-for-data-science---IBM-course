A database is a repository of data, it's a program which storage data. Provides functionality for adding, modifying an querying that data. There are different kind of databases of different requirements.

A relational data base storage data in tabular form - columns and rows. Columns contains item properties about the item, for instance: last name, first name etc.

A table is a collection of related things of related things, for instance: employees, authors.
In a relational data base, you can form relations between tables.

# Relational Database Concepts.

## Relational Model

This model allows for data independence.

Data is stored in tables.

In this case there is logical data independence, physical data independence, physical storage independence.




## Entity relational model

In the ER model, entities are objects that exists independently of any other entities in the database.
The building blocks of an ER diagram are entities and attributes.

An __entity__ can be a noun, person, place or thing. In a __ER diagram, entities are drawn as a rectangle__. 
Entities havee __attributes__ which are the data elements that characterize the entity. 



Attributes tell us more about the entity. In an __ER diagram attributes are drawn as ovals__.



Atributes are connected to exactly one entity.

## Mapping entity diagrams to tables

A table is a combination of rows and columns. While mapping, the entity becomes the table.
Having said that, the table has not yet taken the form of rows and columns. 

The attributes get translated into columns in a table providing the actual table form of rows and columns.
Later, we add some data values to each of the columns, which completes the table form.

Each attribute stores data values of different formats: characters, numbers, dates, currency, etc.
In the book table example, the title is made up of characters. As book titles vary in length, we can set the variable character data type for the title column _Varchar_. The ISBN column would be _Char_ because it contains dashes as well as numbers and so on. 




Using the book entity mapping as an example, we can create the tables for the remainder of our simplified library example using entity names, like author, author list, borrower, loan, and copy.

The entity attibutes will be the columns of the tables. Each table is assigned a primary key. The primary key of a relational table uniquely identifies each tuple or row in a table, preveting duplication of data and providing a way of defining relationships between tables.



Tables also contain foreign keys which are primary keys defined in other tables. 

__Summary (important ideas):__

1. Entitites are independent objects which can have multiple characteristicds called attributes.
2. When mapping to a relational database, entities are represented as tables and attributes map to columns.
3. Common data types include characters, such as _Char_ and _Varchar_, numbers such as integer and decimal, and timestamps including data and time.
4. A primary key uniquely identifies a specific row in a table and prevents duplication of data.
