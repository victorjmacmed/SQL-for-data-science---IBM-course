 ## Example: create table statement.
 
 SQL statements are used for interacting with Entities (that is tables), attributes (that is columns) and their tuples (or rows with data values) in relation databases.
 
 SQL Statements fall into two different categories:
 
 1. Data definition Language Statements 
 2. Data manipulation Language Statements

### Data definition Language (DDL) Statements.

Data definition language statements (DDL) are used to define, change or drop database objects such as tables.
Common DDL statements types include: _CREATE,ALTER,TRUNCATE, and DROP_.

- ALTER: is used for altering tables including adding and dropping columns and modifying their datatypes;

- TRUNCATE: is used for deleting data in a table but no the table itself;

- DROP: is used for deleting tables.

### Data Manipulation Language (DML) Statements.

DML statements are used to read and modify data in tables. These are also sometimes referred to as CRUD operations, that is, Create, Read, Update and Delete rows in a table. Common DML statement types include: INSERT, SELECT, UPDATE and DELETE.

- INSERT: is used for inserting a row or sevel rows of data into a table;

- SELECT: reads or selects row or rows from a table;

- UPDATE: edits row or rows in a table;

- DELETE: removers a row or rows of data from a table.

## Using the SELECT Statement (Retrieving data from a data base table)

After creating a relational database table and inserting data into the table, we want to see the data. To see the data, we use the SELECT statement. The SELECT statement is a data manipulation language or DML statement. DML statements are used to read and modify data.

The SELECT statement is called a __query__ and the output we get from executing this query is called a __result set__ or a __result table__. In its simplest form, a SELECT statement is: 

_Select * from table name_

__Example: book entity example again.__

db2 => select * from Book

This gives as a result four rows.



__Example: select <column1,column2, ... column n from Book

db2 => select book_id, title, edition, year, price, ISBN, pages, aisle, description from Book.



You can retrieve just the columns you want. The order of the columns displayed alway matches the orde in the SELECT statement.

Example: SELECT <column 1>, <column 2> from Book.

db2 => select book_id, title from Book



The order of the columns displayes matches the order of the columns displayed.

__What if we want to know the title of the book whose book_id is B1?__

Relational operations help use in restricting the result set by allowing us to use the clause WHERE.
The WHERE clause alway requires a predicate. A predicate is conditioned evaluates to true, false or unknown.
Predicates are used in the search condition of the WHERE clause. If we need to know the title of the book whose book_id is B1, we use WHERE clause with the predicate book_id equals B1.

Select book_id title from book where book_id equals B1. 



Notice the result set is now restricted to just one row whose condition evaluates to true.

The previous example used comparison operators supported by a relational database management system: equal to, greater than, less than, less than or equal to, greater than or equal to, and not equal to.



## Select statement examples

Using SELECT queries to retrieve data from the database.

The general syntax of SELECT statements is 

_select COLUMN1, COLUMN2, ... FROM TABLE1_

To retrieve all columns from COUNTRY table we could use "*" instead of specifying individual column names.

_select * from COUNTRY_

The WHERE clause can be added to your query to filter results or get specific rows of data. To retrieve data for all rows in the COUNTRY table the ID is less than 5 we write the code:

_select * from COUNTRY where ID <5_

In case of character based columns the values of the predicates in the where clause need to be enclosed in single quotes. To retrieve the data for the country with country code "CA" we write the code:

_select * from COUNTRY where CCODE = "CA"_

## Using the UPDATE and DELETE Statements.

Altering and deleting data in a relationla data base.

The __UPDATE__ statement is one of the DML statements (read and modify data).

Suppose that we want to update the author entity. AUTHOR is the entity and the columns are the attributes.
Now we want to alter or modify the data in this table. To do this we use the UPDATE statement. The syntax for this statement is





__Example__ We want to update the first name and the last name of the with Author_ID A_2 from Rav Ahuja to Lakshmi Katta.



We write the code

UPDATE AUTHOR

SET LASTNAME = 'KATTA'
    FIRSTNAME='LAKSHMI'
   WHERE AUTHOR_ID = 'A2'
   
 __Remark:__ If we don't specify the WHERE clause, all the rows in the table will be updated.
 All the rows in the table would have changed to Rav Ahuja.
 



The __DELETE__ statement is one of DML statements used to read and modify data. We can use the DELETE statement to remove 1 or more rows from a table.The syntax for this statement is



Using again the AUTHOR entity example. Suppose that we want to delete the rows for author ID A2 and A3. Then we write the code

DELETE FROM AUTHOR 

    WHERE AUTHOR_ID IN ('A2','A3')




__Remark:__ If you don't specify the WHERE clause, all the rows in the table will be removed.
